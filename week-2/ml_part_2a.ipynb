```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

Decision Tree(classification)


```python
data=pd.read_csv(r"C:\python\ml\datasets\LogisticRegression.csv")
x=data.iloc[:,:-1]
y=data["Purchased"]

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier(max_depth=5)
dt.fit(x_train,y_train)
print(dt.score(x_test,y_test)*100)
print(dt.score(x_train,y_train)*100)
dt.predict([[41,85000]])

from sklearn.tree import plot_tree
plot_tree(dt)
plt.show()
plt.figure(figsize=(7,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=dt)
plt.show()
```

K Nearest neighbours(CLassification)

```python
data=pd.read_csv(r"C:\python\ml\datasets\LogisticRegression.csv")
x=data.iloc[:,:-1]
y=data["Purchased"]

from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
sc.fit(x)
x=pd.DataFrame(sc.transform(x),columns=x.columns)

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,random_state=42)

from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=2)
knn.fit(x_train,y_train)
print(knn.score(x_test,y_test)*100)
print(knn.score(x_train,y_train)*100)

print(knn.predict([[1.636088,	0.808379]]))
plt.figure(figsize=(8,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=knn)
plt.show()
```

Hyparameter Tuning , GridSearchCV and RandomisedCv

```python
data=pd.read_csv(r"C:\python\ml\datasets\polynomial.csv")
x=data[["Level"]]
y=data["Salary"]

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=42)

from sklearn.tree import DecisionTreeRegressor
dt=DecisionTreeRegressor(criterion='squared_error',splitter='best',max_depth=10)
dt.fit(x_train,y_train)
print(dt.score(x_test,y_test)*100)
print(dt.score(x_train,y_train)*100)

from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
df={ "criterion": ['squared_error', 'friedman_mse', 'absolute_error', 'poisson'] ,
    "splitter":  ['best', 'random'] ,
    "max_depth": [i for i in range(2,20)],}
gd=GridSearchCV(DecisionTreeRegressor(),param_grid=df)

gd.fit(x_train,y_train)
gd.best_score_
gd.best_params_
```

```python
rd=RandomizedSearchCV(DecisionTreeRegressor(),param_distributions=df,n_iter=20)
rd.fit(x_train,y_train)
rd.best_score_
rd.best_params_
```

Unsupervised Learning

K means Clustering

```python
data=pd.read_csv(r"C:\python\ml\datasets\iris.csv")
data=data.drop(columns="species")
sns.pairplot(data=data)
plt.show()
from sklearn.cluster import KMeans
wcss=[]
for i in range(2,21):
    km=KMeans(n_clusters=i,init="k-means++")
    km.fit(data)
    wcss.append(km.inertia_)
plt.figure(figsize=(8,2))
plt.plot([i for i in range(2,21)],wcss)
plt.xticks([i for i in range(2,21)])
plt.xlabel("nofclusters")
plt.ylabel("wcss")
plt.grid(axis='x')
plt.show()
```

```python
kmn=KMeans(n_clusters=3)
data["predict"]=kmn.fit_predict(data)
data
org_data=pd.read_csv(r"C:\python\ml\datasets\iris.csv")
sns.pairplot(data=data,hue="predict")
sns.pairplot(data=org_data,hue="species")
plt.show()

```

Silhouette Score

```python
from sklearn.metrics import silhouette_score
# silhouette_score(data,labels=kmn.labels_)
ss=[]
no_c=[j for j in range(2,21)]
for i in range(2,21):
    kmn=KMeans(n_clusters=i)
    kmn.fit(data)
    ss.append(silhouette_score(data,labels=kmn.labels_))
plt.plot(no_c,ss)
plt.xlabel("n_clusters")
plt.ylabel("silhouette_score")
plt.xticks(no_c)
plt.grid(axis="x")
plt.show()
```
